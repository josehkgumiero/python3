```python
def positive_or_negative(number):
    if number > 0:
        return "Positive!"
    elif number < 0:
        return "Negative!"
    else:
        return "Neither! It is zero"
```

```python
positive_or_negative(10)
```

```python
positive_or_negative(-10)
```

```python
positive_or_negative(0)
```

```python
def calculator(operation, first_number, second_number):
    if operation == "add":
        return first_number + second_number
    elif operation == "subract":
        return first_number - second_number
    elif operation == "multiply":
        return first_number * second_number
    elif operation == "divide":
        return first_number / second_number
    else:
        return "I do not know what you want me to do!"
```

```python
calculator("add", 4, 4)
```

```python
calculator("subract", 4, 4)
```

```python
calculator("multiply",4 , 4)
```

```python
calculator("divide", 4, 4)
```

```python
def count_down_from(number):
    if number <= 0:
        return

    print(number)
    count_down_from(number-1)
```

```python
count_down_from(5)
```

```python
def reverse(str):
    start_index = 0
    last_index = len(str) - 1
    reversed_string = ""

    while last_index >= start_index:
        reversed_string += str[last_index]
        last_index -= 1

    return reversed_string
```

```python
reverse("straw")
```
