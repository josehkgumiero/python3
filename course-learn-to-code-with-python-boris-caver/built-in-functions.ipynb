```python
numbers = [4, 8, 16, 23, 42]
```

```python
cubes = [number ** 3 for number in numbers]
```

```python
def cube(number):
    return number ** 3
```

```python
list(map(cube, numbers))
```

```python
animals = ["elephant", "horse", "cat", "giraffe", "cheetah", "dog"]
```

```python
def is_long_animal(animal):
    return len(animal) > 5
```

```python
list(filter(is_long_animal, animals))
```

```python
metals = ["gold", "silver", "platinum", "palladium"]
```

```python
list(filter(lambda metal: len(metal)> 5, metals))
```
