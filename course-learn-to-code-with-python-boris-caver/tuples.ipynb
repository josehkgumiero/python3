```python
def my_max(nonsense, *numbers):
    print(nonsense)
    greatest = numbers[0]
    for number in numbers:
        if number > greatest:
            greatest = number
    return greatest
        
```

```python
my_max("Shazam", 1)
```

```python
my_max("Hoorah", 1, 3)
```

```python
my_max("Bonanaza", 1, 3, 9, 6, 7, 8, -14)
```

```python
my_max(1, 3, nonsense = "bonanaza")
```
