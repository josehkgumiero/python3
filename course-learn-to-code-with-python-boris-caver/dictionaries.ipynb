```python
def count_words(words):
    counts={}
    for word in words:
        if word in counts:
            counts[word] +=1
        else:
            counts[word] = 1
    return counts
```

```python
words = ["danger","beware","danger","beware","beware"]
```

```python
count_words(words)
```

```python
def collect_keyword_arguments(**wkargs):
    print(kwargs)
    for key, value in kwargs.items():
        print(f"the key is {key} and the value is {value}")
```

```python
collect_keyword_arguments(a=2, b=3, c=4, d=5)
```

```python
def args_and_kwaargs(a,b, *args, **kwargs):
    print(f"sum a + b {a+b}")
    print(f"sum total of args {sum(args)}")
    dict_total = 0
    for value in kwargs.values():
        dict_total += value
    print(f"total of kwargs is {dict_total}")
```

```python
args_and_kwaargs(1,2,3,4,5,6,x=8, y=9, z=10)
```

```python
def height_to_meters(feet, inches):
    total_inches = (feet *12 )+ inches
    return total_inches * 0.0254
```

```python
height_to_meters(5,11)
```

```python
stats = {
    "feet": 5,
    "inches": 11
}
```

```python
height_to_meters(**stats)
```
