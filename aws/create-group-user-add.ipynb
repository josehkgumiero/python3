```python
import boto3
from botocore.exceptions import ClientError
```

```python
# ===== CONFIGURAÇÕES =====
AWS_ACCESS_KEY = 'SUA_ACCESS_KEY'
AWS_SECRET_KEY = 'SUA_SECRET_KEY'
REGION = 'us-east-1'  # IAM é global, mas boto3 exige região

GROUP_NAME = 'grupo-desenvolvedores'
USER_NAME = 'usuario-teste'
```

```python
# ===== CLIENTE IAM =====
iam = boto3.client(
    'iam',
    region_name=REGION,
    aws_access_key_id=AWS_ACCESS_KEY,
    aws_secret_access_key=AWS_SECRET_KEY
)
```

```python
# ===== CRIAR GRUPO =====
try:
    iam.create_group(GroupName=GROUP_NAME)
    print(f" Grupo '{GROUP_NAME}' criado com sucesso!")

except ClientError as e:
    if e.response['Error']['Code'] == 'EntityAlreadyExists':
        print(f" Grupo '{GROUP_NAME}' já existe.")
    else:
        print(" Erro ao criar grupo:")
        print(e.response['Error']['Message'])
```

```python
# ===== CRIAR USUÁRIO =====
try:
    iam.create_user(UserName=USER_NAME)
    print(f" Usuário '{USER_NAME}' criado com sucesso!")

except ClientError as e:
    if e.response['Error']['Code'] == 'EntityAlreadyExists':
        print(f" Usuário '{USER_NAME}' já existe.")
    else:
        print(" Erro ao criar usuário:")
        print(e.response['Error']['Message'])
```

```python
# ===== ADICIONAR USUÁRIO AO GRUPO =====
try:
    iam.add_user_to_group(
        GroupName=GROUP_NAME,
        UserName=USER_NAME
    )
    print(f" Usuário '{USER_NAME}' adicionado ao grupo '{GROUP_NAME}'!")

except ClientError as e:
    print(" Erro ao adicionar usuário ao grupo:")
    print(e.response['Error']['Message'])
```
