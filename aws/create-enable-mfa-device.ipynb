```python
import boto3
from botocore.exceptions import ClientError
```

```python
iam = boto3.client(
    "iam",
    aws_access_key_id="SUA_ACCESS_KEY",
    aws_secret_access_key="SEU_SECRET_KEY",
    region_name="us-east-1"
)
```

```python
usuario = "nome-do-usuario-iam"
mfa_name = f"{usuario}-mfa"
```

```python
try:
    response = iam.create_virtual_mfa_device(
        VirtualMFADeviceName=mfa_name
    )

    mfa_arn = response['VirtualMFADevice']['SerialNumber']
    base32 = response['VirtualMFADevice']['Base32String']

    print("MFA criado com sucesso")
    print("ARN do MFA:", mfa_arn)
    print("Chave Base32 (para app autenticador):", base32)

except ClientError as e:
    print("Erro ao criar MFA:", e)
```

```python
codigo1 = "123456"  # primeiro código do app
codigo2 = "789012"  # segundo código do app
```

```python
try:
    iam.enable_mfa_device(
        UserName=usuario,
        SerialNumber=mfa_arn,
        AuthenticationCode1=codigo1,
        AuthenticationCode2=codigo2
    )

    print("MFA habilitado com sucesso para o usuário.")

except ClientError as e:
    print("Erro ao habilitar MFA:", e)
```
