```python
import boto3
from botocore.exceptions import ClientError

iam = boto3.client(
    'iam',
    aws_access_key_id='SUA_ACCESS_KEY_ID_AQUI',
    aws_secret_access_key='SUA_SECRET_ACCESS_KEY_AQUI'
    #region_name='us-east-1'  # IAM não depende de região, mas evite erro
)

alias_name = "meu-alias-aws"

try:
    iam.create_account_alias(AccountAlias=alias_name)
    print(f"Account Alias '{alias_name}' criado com sucesso!")

except ClientError as e:
    print("Erro ao criar o alias:")
    print(e.response['Error']['Message'])
```

```python
{
  "Effect": "Allow",
  "Action": [
    "iam:CreateAccountAlias",
    "iam:ListAccountAliases",
    "iam:DeleteAccountAlias"
  ],
  "Resource": "*"
}
```
