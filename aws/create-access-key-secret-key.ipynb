```python
import boto3

iam = boto3.client(
    'iam',
    aws_access_key_id='SUA_ACCESS_KEY_ATUAL',
    aws_secret_access_key='SUA_SECRET_ATUAL'
)

usuario_iam = 'nome-do-usuario-iam'

response = iam.create_access_key(
    UserName=usuario_iam
)

print("Access Key ID:")
print(response['AccessKey']['AccessKeyId'])

print("\nSecret Access Key:")
print(response['AccessKey']['SecretAccessKey'])
```
