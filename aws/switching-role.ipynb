```python
import boto3
from botocore.exceptions import ClientError, NoCredentialsError, ParamValidationError
```

```python
# ====== CREDENCIAIS DE ORIGEM ======
ACCESS_KEY = "SUA_ACCESS_KEY"
SECRET_KEY = "SEU_SECRET_KEY"
REGION = "us-east-1"
```

```python
# ====== ROLE A SER ASSUMIDA ======
ROLE_ARN = "arn:aws:iam::123456789012:role/NomeDaRole"
ROLE_SESSION_NAME = "sessao-python-switch-role"
```

```python
iam = boto3.client(
    "iam",
    aws_access_key_id="SUA_ACCESS_KEY",
    aws_secret_access_key="SEU_SECRET_KEY",
    region_name="us-east-1"
)
```

```python
try:
    # Cliente STS com Access Key e Secret Key
    sts_client = boto3.client(
        "sts",
        aws_access_key_id=ACCESS_KEY,
        aws_secret_access_key=SECRET_KEY,
        region_name=REGION
    )

    # Assume a Role
    response = sts_client.assume_role(
        RoleArn=ROLE_ARN,
        RoleSessionName=ROLE_SESSION_NAME
    )

    credentials = response["Credentials"]

    print("Role assumida com sucesso!")

except NoCredentialsError:
    print(" Credenciais AWS não encontradas")

except ParamValidationError as e:
    print(" Erro de parâmetro:", e)

except ClientError as e:
    print(" Erro ao assumir a role:", e)
```
